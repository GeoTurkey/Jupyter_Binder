* For IGS's 30s clock products the correct abbrevation should be "IGS_30"

* The code will also automatically unzip all the compressed (.Z) files extracted from the WEB.

```python

################################# DOWNLOAD PRECISE CLOCKS FROM CDDIS NASA #####################################

import datetime
import requests
import os
import gnsscal
os.chdir("/home/engin/Desktop/Python_GIS/GNSS_DATA")



# These three rows of code must be populated by user!!!!
# Use three digits abbreviation for your AC and use uppercase. (e.g., IGS, IGR, COD, GFZ, JPL, ESA, MIT, COD, COF )
# For CODE's 5s clocks AC = "COD_O5"; For IGS's 30s clocks AC="IGS_30"
################################################

start_date = datetime.date(2015,2,20) 
end_date = datetime.date(2015,2,24)
AC="igs_30"

################################################


if AC=="igs" or AC=="IGS" or AC=="igr" or AC=="IGR" or AC=="cod" or AC=="COD" \
    or AC == "JPL" or AC == "jpl" or AC=="MIT" or AC=="mit" \
    or AC == "GFZ" or AC == "gfz" or AC=="esa" or AC=="ESA" \
    or AC == "GRG" or AC == "grg" or AC == "EMR" or AC == "emr":
        ext=".clk.Z"
        ext2=".clk"

if AC=="COD_05" or AC=="cod_05" or AC=="COF_05" or AC=="cof_05":
        AC=AC[:3]
        ext=".clk_05s.Z"
        ext2=".clk_05s"
        
if AC=="IGS_30" or AC=="igs_30":
        AC=AC[:3]
        ext=".clk_30s.Z"
        ext2=".clk_30s"

url_base = "https://cddis.nasa.gov/archive/gnss/products/"

date_generated = [start_date + datetime.timedelta(days=x) for x in range(0, (end_date-start_date).days+1)]

for date in date_generated:
    gpsweek=gnsscal.date2gpswd(date)[0]
    dofw=gnsscal.date2gpswd(date)[1]
    url=url_base+str(gpsweek)
    prd=AC.lower()+str(gpsweek)+str(dofw)+ext
    prd2=AC.lower()+str(gpsweek)+str(dofw)+ext2
    if os.path.isfile(prd2):
        print('{} file is exist'.format(prd2))
        url=url_base
        continue
    url_file=url+"/"+prd
    r = requests.get(url_file)
    with open(prd, 'wb') as f:
        f.write(r.content)
    os.system("gunzip "+prd)
        print('{} file has been downloaded'.format(prd))
print("All requested files were downloaded...Good Bye...")
```

DOWNLOAD MGEX ORBITS AND CLOCKS FROM CDDIS NASA

```python

##################### DOWNLOAD MGEX ORBITS,CLOCKS, ERP and BIAS FROM CDDIS NASA #####################################

import datetime
import requests
import os
import gnsscal
import gzip
os.chdir("/home/engin/Desktop/Python_GIS/GNSS_DATA")



# These three rows of code must be populated by user!!!!
################################################

start_date = datetime.date(2020,1,1)
end_date = datetime.date(2020,1,1)
AC="COD" # COD, GFZ, GRG, JAX, WUM

################################################

def exists(path):
    r = requests.head(path)
    return r.status_code == requests.codes.ok
    
url_base = "https://cddis.nasa.gov/archive/gnss/products/mgex/"
#https://cddis.nasa.gov/archive/gnss/products/2083/cod20831.eph.Z

date_generated = [start_date + datetime.timedelta(days=x) for x in range(0, (end_date-start_date).days+1)]

for date in date_generated:
    doy=date.strftime("%j")
    year=date.strftime("%Y")
    gpsweek=gnsscal.date2gpswd(date)[0]
    dofw=gnsscal.date2gpswd(date)[1]
    url=url_base+str(gpsweek)
    prd1=AC.upper()+"0MGXFIN_"+year+doy+"0000_01D_05M_ORB.SP3.gz"
    prd2=AC.upper()+"0MGXFIN_"+year+doy+"0000_01D_30S_CLK.CLK.gz"
    prd3=AC.upper()+"0MGXFIN_"+year+doy+"0000_03D_12H_ERP.ERP.gz"
    prd4=AC.upper()+"0MGXFIN_"+year+doy+"0000_01D_01D_OSB.BIA.gz"
    lst=[prd1,prd2,prd3,prd4]
    for n in lst:
        url_file=url+"/"+n
        if exists(url_file)==True:
            r1 = requests.get(url_file)
            with gzip.open(n, 'wb') as f1:
                f1.write(r1.content)
                print('{} file has been downloaded'.format(n))
            os.system("gunzip "+n)
        else:
            print("Either the file {} is missing or not readable".format(n))

print("All requested MGEX files were downloaded...Good Bye...")


```

```python
################################# DOWNLOAD IGS IONOSPHERE IONEX AND CODE ION FILES  #####################################

import datetime
import requests
import os
import gnsscal
import gzip

os.chdir("/home/engin/Desktop/Python_GIS/GNSS_DATA")

# These three rows of code must be populated by user!!!!
################################################
start_date = datetime.date(2020,1,1)
end_date = datetime.date(2020,1,1)
AC="cod"  # cod, cor, esa, esr, igs, igr, jpl, jpr, whu, whr, upc, upr,bern (bernese ion)
############################################################################################


url_base_ionex="https://cddis.nasa.gov/archive/gnss/products/ionex/"


date_generated = [start_date + datetime.timedelta(days=x) for x in range(0, (end_date-start_date).days+1)]

for date in date_generated:
    doy=date.strftime("%j")
    year=date.strftime("%Y")
    yr2=date.strftime("%Y")[2:]
    gpsweek=gnsscal.date2gpswd(date)[0]
    dofw=gnsscal.date2gpswd(date)[1]
    
    if AC=="bern" or AC=="BERN" or AC=="Bern":
        url_base_bern ="http://ftp.aiub.unibe.ch/CODE/"
        url_bern=url_base_bern+year+"/"+"COD"+str(gpsweek)+str(dofw)+".ION.Z"
        name="COD"+str(gpsweek)+str(dofw)+".ION.Z"
        name2="COD"+str(gpsweek)+str(dofw)+".ION"
        if os.path.isfile(name2):
            print('{} file is exist'.format(name2))
            continue
        r = requests.get(url_bern)
        with open(name, 'wb') as f:
            f.write(r.content)
            print('ION file {} has been downloaded.'.format(name, AC))
        os.system("gunzip "+name)
        print("All requested CODE ION files were downloaded and unzipped. Good Bye...")

    else:
        url=url_base_ionex+year+"/"+doy+"/"
        prd=AC.lower()+"g"+doy+"0.20i.Z"
        prd2=AC.lower()+"g"+doy+"0.20i"
        if os.path.isfile(prd2):
            print('{} file is exist'.format(prd2))
            continue
        url_file=url+prd
        r = requests.get(url_file)
        with open(prd, 'wb') as f:
            f.write(r.content)
            print('IONEX file {} from {} has been downloaded.'.format(prd2, AC))
        os.system("gunzip "+prd)
        print("All requested IONEX files were downloaded and unzipped. Good Bye...")
    
```

DOWNLOAD VMF1 OR VMF3 GRID FILES FROM Vienna Mapping Functions Open Access Data

```python

################################# DOWNLOAD VMF GRID FILES #####################################

import datetime
import requests
import os
import gnsscal
import gzip
os.chdir("/home/engin/Desktop/GNSS/gen/VMFGRID/")



# These three rows of code must be populated by user!!!!
#####################################################################################

start_date = datetime.date(2020,2,22)
end_date = datetime.date(2020,2,23)
GRID="VMF1"   # VMF1: 2.5 x 2.0  ;  VMF3: 1.0 x 1.0 
#############################################################################################3

str_date=start_date - datetime.timedelta(days=1)
en_date=end_date + datetime.timedelta(days=1)


    
url_vmf1_op="https://vmf.geo.tuwien.ac.at/trop_products/GRID/2.5x2/VMF1/VMF1_OP/"
url_vmf3_op="https://vmf.geo.tuwien.ac.at/trop_products/GRID/1x1/VMF3/VMF3_OP/"



#https://vmf.geo.tuwien.ac.at/trop_products/GRID/2.5x2/VMF1/VMF1_OP/2020/VMFG_20200101.H00
#https://vmf.geo.tuwien.ac.at/trop_products/GRID/1x1/VMF3/VMF3_OP/2020/VMF3_20200101.H00


date_generated = [str_date + datetime.timedelta(days=x) for x in range(0, (en_date-str_date).days+1)]

for date in date_generated:
    doy=date.strftime("%j")
    year=date.strftime("%Y")
    dt=date.strftime("%Y%m%d")


    
    if GRID=="VMF1" or GRID=="vmf1":
        url=url_vmf1_op + year
        prd1="VMFG_"+dt+".H00"
        prd2="VMFG_"+dt+".H06"
        prd3="VMFG_"+dt+".H12"
        prd4="VMFG_"+dt+".H18"

        prd_list=[prd1,prd2,prd3,prd4]
        
        for grd in prd_list:
            if os.path.isfile(grd):
                print('{} file is exist'.format(grd))
                url=url_vmf1_op
                continue
            url_file=url+"/"+grd
            r1 = requests.get(url_file)
            with open(grd, 'wb') as f1:
                f1.write(r1.content)
                print('{} grid file has been downloaded'.format(grd))
        
    if GRID=="VMF3" or GRID=="vmf3":
        url=url_vmf3_op + year
        prd1="VMF3_"+dt+".H00"
        prd2="VMF3_"+dt+".H06"
        prd3="VMF3_"+dt+".H12"
        prd4="VMF3_"+dt+".H18"

        prd_list=[prd1,prd2,prd3,prd4]
        
        for grd in prd_list:
            if os.path.isfile(grd):
                print('{} file is exist'.format(grd))
                url=url_vmf3_op
                continue
            url_file=url+"/"+grd
            r1 = requests.get(url_file)
            with open(grd, 'wb') as f1:
                f1.write(r1.content)
                print('{} grid file has been downloaded'.format(grd))
        
    
print("All requested VMF grid files were downloaded...Good Bye...")
```
